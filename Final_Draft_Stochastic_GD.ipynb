```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.array([1,3,5]) # Feature
y_actual= np.array([5,12,18]) # Actual Data Point

w0 = 0 # weight of intersept
w1 = 0 # weight of slope
alpha = 0.095 # step 
g,h = [],[]

for Epoch in range(10):
    g.append(Epoch)
    a,b,c,d = [],[],[],[]
    
    for r in range(len(x)):
        
        '''Predicted Data Point'''
        y_predicted = w0 + w1*x
        a.append(y_predicted[r])
        j.append(y_predicted[r])
        
        error = y_predicted - y_actual
        b.append(error[r])
        k.append(error[r])
        
        error_x = error*x
        c.append(error_x[r])
        
        s_e = error**2
        d.append(s_e[r])
        
        w0 -= alpha*error[r]
        w1 -= alpha*error_x[r]

    
    '''Proper representation of data is in pandas'''
    Table = list(zip(x,y_actual,a,b,c,d)) 
    df = pd.DataFrame(Table)
    df.set_axis(['x','y','y_predicted', 'error', 'error_x', 'squares error'], axis=1, inplace = True) 
    print(df)
    
    L = sum(d)/len(x) # L loss function
    h.append(L)
    '''Sum of errors and Sum of squares of errors '''
    
    print(f'Sum of errors = {sum(b)} \nSum of squares of errors = {sum(d)}')
    
    '''Loss function which is to minimized'''
    
    print(f'Value of Loss Function after {Epoch+1} Epoch  = {L}')
    
    '''New updated weights for intercept and slopes'''
    
    print(f'Updated value for w0 = {w0} \nUpdated value of w1 = {w1} ') 
```
